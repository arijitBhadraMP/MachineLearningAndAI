```python
import nltk
from nltk.corpus import state_union
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
```

```python
var01 = state_union.raw('/usr/local/arijit/Work_Asheesh/TAK.txt')
```

```python
var01
```

```python
import re
var02=re.sub('\d+','',var01)
var02
```

```python
var03=var02.lower()
```

```python

tokenizer = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(var03)
filtered_words = [w for w in tokens if not w in stopwords.words('english')]

filtered_words
```

```python
import csv

with open('new', 'w') as myfile:
    wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
    wr.writerow(['SNo'])
    wr.writerow(filtered_words)
```
