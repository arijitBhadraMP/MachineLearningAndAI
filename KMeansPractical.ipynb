#### k-Means Clustering
   * Unsupervised Machine Learning
   * Searches for a pre-determined number of clusters within an unlabeled multidimensional dataset
   * The centroid/custer center is mean of all the points belonging to the cluster



```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
X, y_true = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50);
#plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50)
```


#### k-means

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)       # The number of clusters to form as well as the number of centroids to generate.
kmeans.fit(X)                       # Compute k-means clustering(making the object)
y_kmeans = kmeans.predict(X)        # Predict the closest cluster each sample in X belongs to.
kmeans.cluster_centers_             # Coordinates of cluster centers
```

```python
y_kmeans
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

#### k-means on digits

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.data.shape
```

```python
digits.data #We have this
```

```python
kmeans = KMeans(n_clusters=10, random_state=0)
```

```python
clusters = kmeans.fit_predict(digits.data) #Compute cluster centers and predict cluster index for each sample.
                                          # Here, predicting which cluster the data belongs to
```

```python
clusters.shape
```

```python
clusters
```

```python
kmeans.cluster_centers_
```

```python
ig, ax = plt.subplots(2, 5, figsize=(8, 3))
centers = kmeans.cluster_centers_.reshape(10, 8, 8)
for axi, center in zip(ax.flat, centers):
    axi.set(xticks=[], yticks=[])
    axi.imshow(center, interpolation='nearest', cmap=plt.cm.binary)
```
