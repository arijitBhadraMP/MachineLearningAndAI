## K Means

```python
from sklearn.cluster import KMeans
```

```python
import pandas as pd
```

```python
cost = []
models = []
for i in range(1,20):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit((diabetes_data.as_matrix()[:,2:]))
    models.append(kmeans)
    print (kmeans.score((diabetes_data.as_matrix()[:,2:])))
    cost.append(-1 * kmeans.score((diabetes_data.as_matrix()[:,2:])))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Idendifying the number of Clusters

```python
plt.plot(range(1,20), cost)
```

As we can see from above drastical change stops and creates an "elbow" in 4; so we are considering 4 as the number of models;

```python
kmeans = KMeans(n_clusters=4, random_state=0)
```

```python
kmeans.fit(diabetes_data.as_matrix()[:,2:])
```

```python
class_diabetes = kmeans.predict(diabetes_data.as_matrix()[:,2:])
```

#### Check all 4 clusters

```python
diabetes_data.iloc[np.where(class_diabetes == 0)]
```
